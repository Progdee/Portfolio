Use the subplot method to add multiple plots in one figure.
It takes 3 arguments nrows, ncols and index

```python
fig, ax = plt.subplots()
print(fig,ax)
```

```python
plt.subplot(1,2,1)
plt.plot(data.work_force,data.income, "go")
plt.title("Income vs Work Force")
plt.subplot(1,2,2)
plt.plot(data.hospital_beds, data.income, "r^")
plt.title("Income vs Hospital Beds")
plt.suptitle('Sub Plots')
plt.show()
```

```python
fig, ax = plt.subplots(nrows= 2,ncols = 2, figsize = (6,6))
ax[0,0].plot(data.work_force, data.income,'go')
ax[0,1].plot(data.work_force, data.income, "go")

plt.show()
```
